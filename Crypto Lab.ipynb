# 01	Write a program to implement encryption and decryption using Caesar cipher.

```python
def caesar_encrypt(plaintext, shift):
    encrypted_text = ""
    for char in plaintext:
        if char.isalpha():
            if char.islower():
                encrypted_text += chr((ord(char) + shift - ord('a')) % 26 + ord('a'))
                                         # m=109  + 134 - 97=146 % 26 =16 + 97 = 113 = q
            else:
                encrypted_text += chr((ord(char) + shift - ord('A')) % 26 + ord('A'))
        else:
            encrypted_text += char
    return encrypted_text

def caesar_decrypt(ciphertext, shift):
    decrypted_text = ""
    for char in ciphertext:
        if char.isalpha():
            if char.islower():
                decrypted_text += chr((ord(char) - shift - ord('a')) % 26 + ord('a'))
            else:
                decrypted_text += chr((ord(char) - shift - ord('A')) % 26 + ord('A'))
        else:
            decrypted_text += char
    return decrypted_text


plaintext = "meet me after toga party";
print(f"Caesar Original text    : {plaintext}")
shift = 134
ciphertext = caesar_encrypt(plaintext, shift)
print(f"Caesar Cipher Encryption: {ciphertext}")
plaintext=caesar_decrypt(ciphertext,shift)
print(f"Caesar Cipher Decryption: {plaintext}")
```

# 03	Write a program to implement encryption and decryption using Playfair cipher.

```python
def playfair_cipher(plaintext, key, mode):
    # Define the alphabet, excluding 'j'
    alphabet = 'abcdefghiklmnopqrstuvwxyz'
    # Remove whitespace and 'j' from the key and convert to lowercase
    key = key.lower().replace(' ','').replace('j','i')
   
    key_square = ''
    for letter in key + alphabet:
        if letter not in key_square:
            key_square += letter
    # Split the plaintext into digraphs
    plaintext = plaintext.lower().replace(' ','').replace('j','i')
    if len(plaintext) % 2 == 1:
        plaintext += 'x'
    digraphs = [plaintext[i:i+2] for i in range(0, len(plaintext), 2)]

    def encrypt(digraph):
        a, b = digraph
        row_a, col_a = divmod(key_square.index(a), 5)
        row_b, col_b = divmod(key_square.index(b), 5)
        if row_a == row_b:
            col_a = (col_a + 1) % 5
            col_b = (col_b + 1) % 5
        elif col_a == col_b:
            row_a = (row_a + 1) % 5
            row_b = (row_b + 1) % 5
        else:
            col_a, col_b = col_b, col_a
        return key_square[row_a*5+col_a] + key_square[row_b*5+col_b]

    def decrypt(digraph):
        a, b = digraph
        row_a, col_a = divmod(key_square.index(a), 5)
        row_b, col_b = divmod(key_square.index(b), 5)
        if row_a == row_b:
            col_a = (col_a - 1) % 5
            col_b = (col_b - 1) % 5
        elif col_a == col_b:
            row_a = (row_a - 1) % 5
            row_b = (row_b - 1) % 5
        else:
            col_a, col_b = col_b, col_a
        return key_square[row_a*5+col_a] + key_square[row_b*5+col_b]
    
    result = ''
    for digraph in digraphs:
        if mode == 'encrypt':
            result += encrypt(digraph)
        elif mode == 'decrypt':
            result += decrypt(digraph)

    return result

plaintext = 'we are happy today'
key = 'ICE'
print('The Original text   :',plaintext)
ciphertext = playfair_cipher(plaintext, key, 'encrypt')
print('The Encryption text :',ciphertext) 
decrypted_text = playfair_cipher(ciphertext, key, 'decrypt')
print('The Decryption text :',decrypted_text) 
```

# 04	Write a program to implement encryption and decryption using Hill cipher.

```python
import numpy as np

def matrix_mod_inverse(matrix, modulus):
    det = int(np.linalg.det(matrix))
    det_inv = pow(det, -1, modulus)  # Modular inverse of determinant

    adj_matrix = np.round(det_inv * np.linalg.inv(matrix)).astype(int)
    inverse_matrix = (det * adj_matrix) % modulus

    return inverse_matrix

def matrix_to_numbers(matrix, alphabet):
    return [alphabet[num % len(alphabet)] for num in matrix.flatten()]

def numbers_to_matrix(numbers, shape, alphabet):
    return np.array([alphabet[num % len(alphabet)] for num in numbers]).reshape(shape)

def encrypt(plaintext, key, alphabet):
    plaintext = [ord(char) - ord('A') for char in plaintext]
    key_matrix = np.array(key).reshape(len(key), -1)

    while len(plaintext) % len(key) != 0:
        plaintext.append(ord('X') - ord('A'))

    plaintext_matrix = np.array(plaintext).reshape(-1, len(key))
    ciphertext_matrix = np.dot(plaintext_matrix, key_matrix) % len(alphabet) #multiply

    return ''.join(matrix_to_numbers(ciphertext_matrix, alphabet))

def decrypt(ciphertext, key, alphabet):
    ciphertext = [ord(char) - ord('A') for char in ciphertext]
    key_matrix = np.array(key).reshape(len(key), -1)

    ciphertext_matrix = np.array(ciphertext).reshape(-1, len(key))
    inverse_key_matrix = matrix_mod_inverse(key_matrix, len(alphabet))

    plaintext_matrix = np.dot(ciphertext_matrix, inverse_key_matrix) % len(alphabet)

    return ''.join(matrix_to_numbers(plaintext_matrix, alphabet))

def main():
    alphabet = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    key = [[2,3], [3,6]] 

    plaintext = "NAIM"
    encrypted_text = encrypt(plaintext, key, alphabet)
    decrypted_text = decrypt(encrypted_text, key, alphabet)

    print("Original Text:", plaintext)
    print("Encrypted Text:", encrypted_text)
    print("Decrypted Text:", decrypted_text)

if __name__ == "__main__":
    main()
```

# 05	Write a program to implement encryption and decryption using Poly-Alphabetic cipher.

```python
def vigenere_encrypt(plaintext, key):
    encrypted_text = ""
    key_repeated = (key * (len(plaintext) // len(key))) + key[:len(plaintext) % len(key)]
                    #{( 3 * (5//3=1)) + key[:len(plaintext) % len(key)] = "KEY[:5 % 3] = "KEY[2]" = "KE"}=KEY" + "KE" = "KEYKE
    for i in range(len(plaintext)):
        char = plaintext[i].upper()
        if char.isalpha():
            encrypted_char = chr((ord(char) + ord(key_repeated[i]) - 2 * ord('A')) % 26 + ord('A'))
                                       # 72 + 75 = 147 - 2*65 =17 % 26 =17 + 65= 82=R
            encrypted_text += encrypted_char
        else:
            encrypted_text += char

    return encrypted_text

def vigenere_decrypt(ciphertext, key):
    decrypted_text = ""
    key_repeated = (key * (len(ciphertext) // len(key))) + key[:len(ciphertext) % len(key)]

    for i in range(len(ciphertext)):
        char = ciphertext[i].upper()
        if char.isalpha():
            decrypted_char = chr((ord(char) - ord(key_repeated[i]) + 26) % 26 + ord('A'))
            decrypted_text += decrypted_char
        else:
            decrypted_text += char

    return decrypted_text

def main():
    plaintext = "HELLO"
    key = "KEY"

    encrypted_text = vigenere_encrypt(plaintext, key)
    print("Plaintext     :", plaintext)
    print("Encrypted Text:", encrypted_text)

    decrypted_text = vigenere_decrypt(encrypted_text, key)
    print("Decrypted Text:", decrypted_text)

if __name__ == "__main__":
    main()
```

# 06	Write a program to implement encryption and decryption using Vernam cipher.

```python
import random

def generate_key(length):
    return ''.join(random.choice('ABCDEFGHIJKLMNOPQRSTUVWXYZ') for _ in range(length))

def vernam_encrypt(plaintext, key):
    encrypted_text = ''
    for i in range(len(plaintext)):
        char = plaintext[i].upper()
        if char.isalpha():
            encrypted_char = chr((ord(char) + ord(key[i]) - 2 * ord('A')) % 26 + ord('A'))
            encrypted_text += encrypted_char
        else:
            encrypted_text += char
    return encrypted_text

def vernam_decrypt(ciphertext, key):
    decrypted_text = ''
    for i in range(len(ciphertext)):
        char = ciphertext[i].upper()
        if char.isalpha():
            decrypted_char = chr((ord(char) - ord(key[i]) + 26) % 26 + ord('A'))
            decrypted_text += decrypted_char
        else:
            decrypted_text += char
    return decrypted_text

def main():
    plaintext = "HELLO"
    key = generate_key(len(plaintext))

    encrypted_text = vernam_encrypt(plaintext, key)
    print("Plaintext:", plaintext)
    print("Key:", key)
    print("Encrypted Text:", encrypted_text)

    decrypted_text = vernam_decrypt(encrypted_text, key)
    print("Decrypted Text:", decrypted_text)

if __name__ == "__main__":
    main()
```

# 08	Write a program to implement encryption and decryption using Rail Fence cipher.

```python
def rail_fence_encrypt(plaintext, key):
    fence = ['' for _ in range(key)]
    direction = 1  # 1 for downward, -1 for upward
    row = 0

    for char in plaintext:
        fence[row] += char
        row += direction

        if row == key - 1 or row == 0:
            direction *= -1  # Change direction when reaching the top or bottom row

    ciphertext = ''.join(fence)
    return ciphertext

def rail_fence_decrypt(ciphertext, key):
    fence = ['' for _ in range(key)]
    direction = 1
    row = 0

    for char in ciphertext:
        fence[row] += 'X'  # Placeholder for the actual character
        row += direction

        if row == key - 1 or row == 0:
            direction *= -1

    idx = 0
    for i in range(key):
        for j in range(len(fence[i])):
            fence[i] = fence[i][:j] + ciphertext[idx] + fence[i][j+1:]
            idx += 1

    plaintext = ''
    row = 0
    direction = 1

    for _ in range(len(ciphertext)):
        plaintext += fence[row][0]
        fence[row] = fence[row][1:]
        row += direction

        if row == key - 1 or row == 0:
            direction *= -1

    return plaintext

def main():
    plaintext = "HELLO WORLD"
    key = 3

    ciphertext = rail_fence_encrypt(plaintext, key)
    print("Plaintext:", plaintext)
    print("Encrypted Text:", ciphertext)

    decrypted_text = rail_fence_decrypt(ciphertext, key)
    print("Decrypted Text:", decrypted_text)

if __name__ == "__main__":
    main()
```
